### feature engineering 
      is the process of converting raw data into meaning ful features so that a ml model can understand patterns and make better predictions

### 1.feature exploration 
     -the process of analyzing and visualizing features to uncover patterns distribution correlation and potential issues in the 
     dataset before applying ml

### 2.feature creation
       -the process of generating new meaningful features from existing data using domain knowledge or mathematical operations
        to help the model earn better patterns.

### 3.feature encoding
      -the process of converting categorical or non-numeric features into numeric form so machine learning models can process 
       them effectively

### 4.feature scaling
      -the process of transformong numerical features to a standard range or distribution so that no feature dominates
       the learing process due to its scale

### 5.feature selection
        - the process of identifying and keeping only the most relevant features while removing irrelevant or redundant ones
          to improve model performance and reduce overfitting

### feature creation(adds intelligence):

### definition:
      -feature creation is the process of creating new features from existing data using domain knowledge and mathematical 
       operations to help the model learn better patterns

### why it "adds intelligence:
       * raw data may not show hidden relationships
       * new features provide extra information
       * model cannot think logically-feature guide learing

### common types of feature creation:
       * domain-based features(real-world logic)
       * mathematical combinations(sum, average, ratio)
       * interaction features(combinedeffect of features)
       * polunomial features(non-linear relationships)

### example:
    * tota_marks = math+science
    * efficiency = total_marks/study hours

### impact on model:
     * improves accuracy
     * helps models capture complex patterns
     * makes simple models perform better

### feature creation

```python
import pandas as pd

df = pd.DataFrame({
    "Study_Hours":[2,4,6,8,10],
    "Attendence":[60,70,80,90,95],
    "Maths":[40,55,65,75,80],
    "Science":[42,58,68,78,88]
})
```

```python
df
```

### domain-based feature

```python
df["Total_Marks"] = df["Maths"] + df["Science"]
df
```

### mathematical feature

```python
df["Marks_per_hour"] = df["Total_Marks"] / df["Study_Hours"]
df
```

### interaction feature

```python
df["Study_Attendence_Interaction"] = df["Study_Hours"]*df["Attendence"]
df
```

### polynomial feature

```python
df["Study_Hours_Squared"] = df["Study_Hours"]**2
df
```

### feature selection(remove noise):
     definition:
     - feature selection is the process of selecting only relavant features and removing irrelevant or reduntant ones
       to improve model performance

### why it "remove noise":
    -too many features confuse the model
    -irrelevant featres cause overfitting
    -noise reduces prediction accuracy

### types of feature selection:
       - filter methods(correlation,chi-square)
       - wrapper methods(forward,backward,rfe)
       - embedded(lasso,tree-based models)

### example:
     -removing weakly correlated features
     -dropping features with zero importance
     -selecting to features using RFE(recursive feature elimination)

### impact on model:
      -reduce overfitting
      -improves generalization
      -speedsup training
