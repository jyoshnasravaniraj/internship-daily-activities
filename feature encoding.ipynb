# future encoding

feature encoding is the process of converting non-numerical(or) categorical data into a numerical format that machine learning algorithms
can understand process
------->we choose the encoding technique based on:
-the type of categorical data(nominal or ordinal)
-the number of unique in a categorical column,
-and the model we plan to train


# types categorical data

```python
1.nominal data
---no order
2.ordinal data
---ordwer exists
example:
---size(small<medium<large)
---rating(low<medium<large)
```

# why feature encoding is important:

---text data cant be used directly in model training.
---proper encoding
   ---improves model accuracy
   ---prevents wrong assumptions
   ---makes training stable

# common feature encoding techniques

```python
1.label encoding
2.one-hot encoding
3.target encoding
4.ordinal encoding
5.frequency encoding
6.binary encoding
7.feature hashmap encoding
```

# 1.label encoding

---each catrgory is assigned a unique integer(e.g:"red":0,"green":).


# read the data

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "colour": ["red","green","blue","green","red"],
    "size": ["small","medium","large","medium","large"],
    "city": ["hyderabad","delhi","mumbai","delhi","chennai"],
    "target": [10,20,15,25,12]
})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df["colour_label"] = LabelEncoder().fit_transform(df["colour"])
df[["colour","colour_label"]]
```

# one-hot encoding

---creates a new binary column for each category with a 1 indicates the presence of that category and 0 otherwise.


```python
one_hot = pd.get_dummies(df["colour"],prefix="colour")
one_hot
```

```python
ohe = pd.get_dummies(df,columns=["colour"])
print(ohe)
```

# ordinal encoding

---similar to label encoding, but the integer assingment is based on the specific predefined order or rank of the categories(e.g:"low":0,"medium":1,"high":2)

```python
from sklearn .preprocessing import OrdinalEncoder
```

```python
order = [["small","medium","large"]]
df["size_Ordinal"] = OrdinalEncoder(categories=order).fit_transform(df[["size"]])

df[["size","size_Ordinal"]]
```

# target encoding

```python
-each category is replaced with the "mean" of the target variable for that category

```

```python
import pandas as pd
te = df.groupby("city")["target"].mean()
df["city_Target_Enc"] = df["city"].map(te)
df[["city","city_Target_Enc"]]
```

# frequency encoding

```python
    Replaces each category with its frequency(count) in the dataset.
        Useful when the frequencey of a category is predictive of the target variable.
```

```python
fe=df.groupby("city")["target"].count()
df["city_Target_Enc"]=df["city"].map(fe)  #method 1
df[["city","city_Target_Enc"]]


```

```python
f=df["city"].value_counts()
df["city_Target_Enc"]=df["city"].map(fe) # method 2
df[["city","city_Target_Enc"]]
```
